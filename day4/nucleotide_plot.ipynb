```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from collections import Counter
```

```python
# Simulate a DNA sequence
dna_sequence = "ATGCGTACGTTAGGCTAAGCGT"
counts = Counter(dna_sequence)

plt.bar(counts.keys(), counts.values(), color='navy')
plt.xlabel("Nucleotide")
plt.ylabel("Frequency")
plt.title("Nucleotide Frequency in DNA Sequence")
plt.show()

def gc_content(sequence):
    return (sequence.count("G") + sequence.count("C")) / len(sequence) * 100

# Simulated sliding window
window_size = 5
sequence = "ATGCGTACGTTAGGCTAAGCGT"
gc_values = [gc_content(sequence[i:i+window_size]) for i in range(len(sequence) - window_size + 1)]

plt.plot(range(len(gc_values)), gc_values, marker='o', color='green')
plt.xlabel("Position in Sequence")
plt.ylabel("GC Content (%)")
plt.title("GC Content in Sliding Window")
plt.show()
```

```python
# Sample gene expression data for 3 conditions
data = pd.DataFrame({
    "Condition": np.repeat(["Control", "Treatment1", "Treatment2"], 100),
    "Expression": np.concatenate([
        np.random.normal(10, 2, 100),
        np.random.normal(12, 2, 100),
        np.random.normal(14, 2, 100)
    ])
})

sns.boxplot(x="Condition", y="Expression", data=data, palette="pastel")
plt.title("Gene Expression Levels by Condition")
plt.show()
```
